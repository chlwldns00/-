```python
import json
import appdirs
import requests
import pandas as pd



# API 엔드포인트 URL
url = "https://localpay-806adcefc827.herokuapp.com/summaries.json"


# GET 요청 보내기
response = requests.get(url,params={'page':10})

# 응답 상태 코드 확인
if response.status_code == 200:
    # JSON 데이터 출력
    response_data=response.json()
    summaries_list = response_data['summaries']
    main_data = []
    statistics_data = []

    for summary in summaries_list:
        main = {key: summary[key] for key in summary if key != 'statistics'}
        ccdata={**main}
        main_data.append(ccdata)
        for stat in summary['statistics']:
            combined_data = {**stat}
            statistics_data.append(combined_data)

    df_summary_2023 = pd.DataFrame(main_data)
    df_statistics = pd.read_excel('output.xlsx', sheet_name='Statistics2019~2023')


    # with pd.ExcelWriter('output.xlsx') as writer:
    #     df_summary_2023.to_excel(writer, sheet_name='Summary', index=False)
    #     df_statistics.to_excel(writer, sheet_name='Statistics2019~2023', index=False)

    print(df_summary_2023)
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
# 연도별 발행 금액
yearly_issue_amount = df_statistics.groupby('period')['issue_amount'].sum()

# 연도별 사용자 수
yearly_users = df_statistics.groupby('period')['number_of_users'].sum()

# 연도별 가맹점 수
yearly_stores = df_statistics.groupby('period')['number_of_stores'].sum()

# 최대 환전 금액
max_exchange_amount = df_statistics.groupby('period')['exchange_amount_max'].max()

# 최대 캐시백 비율
max_cashback_rate = df_statistics.groupby('period')['cashback_rate_max'].max()

# 지역별 발행 금액
local_issue_amount = df_statistics.groupby('local_metro_name')['issue_amount'].sum()

# 지역별 사용자 수
local_users = df_statistics.groupby('local_metro_name')['number_of_users'].sum()

# 사용자당 평균 발행 금액
df_statistics['average_issue_per_user'] = df_statistics['issue_amount'] / df_statistics['number_of_users']

# 결과 출력
print("Yearly Issue Amount:")
print(yearly_issue_amount)
print("\nYearly Number of Users:")
print(yearly_users)
print("\nYearly Number of Stores:")
print(yearly_stores)
print("\nMax Exchange Amount:")
print(max_exchange_amount)
print("\nMax Cashback Rate:")
print(max_cashback_rate)
print("\nLocal Issue Amount:")
print(local_issue_amount)
print("\nLocal Number of Users:")
print(local_users)
print("\nAverage Issue Amount per User:")
print(df_statistics[['period', 'average_issue_per_user']])
```
