### 5번문제: 타이틀별 댓글 가져와서 출력하기

- 크롤링시 웹페이지 html 호출횟수가 너무 많아지므로 상위 5개의 게시물만 호출하기

````python
import requests
from bs4 import BeautifulSoup
import re

res = requests.get('https://davelee-fun.github.io/trial/board/news.html') #웹사이트 get request
soup = BeautifulSoup(res.content, 'html.parser') #html 파싱
items = soup.select_one('div.list_content') 
link=items.select('a.list_subject')
title= items.select('span.subject_fixed')
numbers_of_replies= items.select('span.rSymph05')
if len(title)==len(numbers_of_replies):
    for i in range(0,5):
        np=int(numbers_of_replies[i].get_text())
        links=link[i] # 제목을 눌렀을때 나오는 본문 페이지 링크
        print(i+1, title[i].get_text().strip(), numbers_of_replies[i].get_text().strip())
        res_page=requests.get('https://davelee-fun.github.io/trial/board/'+links['href']) #
        soup_page=BeautifulSoup(res_page.content, 'html.parser')
        item=soup_page.select_one('div.post_comment')
        item=item.select('div.comment_view')
        #print(item)
        for j in range(0,len(item)):
            context=item[j].get_text().strip().replace('\n','')
            context=re.sub(r'\s+', ' ', context) ## 모든 공백을 띄어쓰기 문자로 대체 하는 정규식
            print('ㄴ',context )
````

### 6번문제: 크롤링 데이터 , 엑셀로 저장하기
- No, 게시글제목, 댓글 로 게시글 제목과 댓글 수 를 저장하기
- 그 다음 라인부터는 해당 게시글의 댓글을 댓글 셀에 저장하기
- 최초 10개의 게시글에 대해서만 작성해보기 (공지 포함)


````python
import requests, openpyxl
from bs4 import BeautifulSoup
import re

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = '게시글'

header = ['No', '게시글제목', '댓글']
excel_sheet.append(header)
excel_sheet.column_dimensions['A'].width = 20
excel_sheet.column_dimensions['B'].width = 80
excel_sheet.column_dimensions['C'].width = 100

res = requests.get('https://davelee-fun.github.io/trial/board/news.html') #웹사이트 get request
soup = BeautifulSoup(res.content, 'html.parser') #html 파싱
items = soup.select_one('div.list_content') 
link=items.select('a.list_subject')
title= items.select('span.subject_fixed')
numbers_of_replies= items.select('span.rSymph05')

if len(title)==len(numbers_of_replies):
    for i in range(0,5):
        np=int(numbers_of_replies[i].get_text())
        links=link[i] # 제목을 눌렀을때 나오는 본문 페이지 링크
       
            
        a=[i+1, title[i].get_text().strip(), numbers_of_replies[i].get_text().strip()]
        excel_sheet.append(a)
        res_page=requests.get('https://davelee-fun.github.io/trial/board/'+links['href']) #
        soup_page=BeautifulSoup(res_page.content, 'html.parser')
        item=soup_page.select_one('div.post_comment')
        item=item.select('div.comment_view')
        #print(item)
        for j in range(0,len(item)):
            context=item[j].get_text().strip().replace('\n','')
            context=re.sub(r'\s+', ' ', context)
            a=['','',context]
            excel_sheet.append(a)

excel_file.save('6번문제.xlsx')
excel_file.close()
        
      
            
````

#### 사용할 주요 openpyxl 기능 

- **Workbook 생성**: `openpyxl.Workbook()`으로 새 Workbook 생성

    ```python
    workbook = openpyxl.Workbook()
    ```

- **Worksheet 이름 변경**: `worksheet.title = 'New Title'`로 Worksheet 이름 변경

    ```python
    worksheet = workbook.active
    worksheet.title = 'Sheet1'
    ```

- **데이터 추가**: `worksheet.append([data1, data2, data3, ...])`로 새 행 추가

    ```python
    worksheet.append(['Value 1', 'Value 2', 'Value 3'])
    ```

- **셀 너비 설정**: `worksheet.column_dimensions['A'].width = 20`로 특정 열 너비 설정

    ```python
    worksheet.column_dimensions['A'].width = 20
    ```

- **폰트 설정**: `openpyxl.styles.Font`로 폰트 속성(굵기, 색상, 밑줄 등) 설정

    ```python
    from openpyxl.styles import Font

    font = Font(bold=True, color='FF0000', underline='single')
    cell.font = font
    ```

- **셀에 하이퍼링크 추가**: `cell.hyperlink = 'https://example.com'`로 특정 셀에 하이퍼링크 추가

    ```python
    cell.hyperlink = 'https://example.com'
    ```

- **셀 정렬 설정**: `openpyxl.styles.Alignment`로 셀 정렬(중앙, 왼쪽, 오른쪽, 위, 아래 등) 설정

    ```python
    from openpyxl.styles import Alignment

    alignment = Alignment(horizontal='center', vertical='top')
    cell.alignment = alignment
    ```

- **텍스트 줄 바꿈**: `Alignment(wrap_text=True)`로 셀 내의 텍스트 자동 줄 바꿈 설정

    ```python
    from openpyxl.styles import Alignment

    alignment = Alignment(wrap_text=True)
    cell.alignment = alignment
    ```

- **셀 테두리 설정**: `openpyxl.styles.Border`와 `openpyxl.styles.Side`로 셀 테두리 스타일 설정

    ```python
    from openpyxl.styles import Border, Side

    # 얇은 선 테두리 스타일
    thin_border = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))
    cell.border = thin_border

    # 굵은 선 테두리 스타일
    thick_border = Border(left=Side(style='thick'), right=Side(style='thick'), top=Side(style='thick'), bottom=Side(style='thick'))
    cell.border = thick_border
    ```

- **셀 배경색 설정**: `openpyxl.styles.PatternFill`로 셀 배경색 설정

    ```python
    from openpyxl.styles import PatternFill

    fill = PatternFill(start_color='FF0000', end_color='FF0000', fill_type='solid')
    cell.fill = fill
    ```

### 7번문제: 6번의 엑셀 시트에,  자동 줄바꿈과 기사제목 볼드체 폰트 적용하기 + 기사 제목에 링크 넣기 + 링크를 가진 게시글 제목 색상과 밑줄 넣기 + 각 셀에 border (선) 넣고, 게시글제목 행 색상 넣기

````python
import requests, openpyxl
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment, Font

wrap_alignment = Alignment(wrap_text=True)
bold_font = Font(bold=True)
header_font = Font(bold=True, size=14)

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = '게시글'

header = ['No', '게시글제목', '댓글']
excel_sheet.append(header)
for cell in excel_sheet[1]:
    cell.font = header_font
    
excel_sheet.column_dimensions['A'].width = 20
excel_sheet.column_dimensions['B'].width = 80
excel_sheet.column_dimensions['C'].width = 80


res = requests.get('https://davelee-fun.github.io/trial/board/news.html')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('div.list_item')
index = 0
row = 2
for item in items[:5]:
    title = item.select_one('span.subject_fixed')
    reply_count = item.select_one('span.rSymph05')
    link = item.select_one('a.list_subject')
    if title != None:
        index += 1
        data1 = [index, title.get_text().strip(), '[' + reply_count.get_text() + ']']
        excel_sheet.append(data1)
        for cell in excel_sheet[row]:
            cell.alignment = wrap_alignment
            cell.font = bold_font
        row += 1
        
        res_title = requests.get('https://davelee-fun.github.io/trial/board/' + link['href'])
        soup_title = BeautifulSoup(res_title.content, 'html.parser')
        replies = soup_title.select('div.comment_content')
        for reply in replies:
            data2 = ['', '', reply.get_text().strip().replace("\n", "").replace("\t", "")]
            excel_sheet.append(data2)
            for cell in excel_sheet[row]:
                cell.alignment = wrap_alignment
            row += 1

excel_file.save('7번문제.xlsx')
excel_file.close()
````
